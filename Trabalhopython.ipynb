```python
import pickle

armazenamento = []

def cadastroarmazenamento(nome, produto, valor):
  armazenamento.append([nome, produto, valor])

while True:
  print("Digite 1 caso queira cadastrar uma nova compra")
  print("Digite 2 caso queira ver os dados inseridos na lista")
  print("Digite 3 caso queira sair do programa")



  opcao = int(input("Digite a opção que deseja: "))

  if opcao == 1:
    nome = input("Digite seu nome: ")
    produto = input("Digite sua compra: ")
    valor = float(input("Digite o valor da compra: "))
    cadastroarmazenamento(nome, produto, valor)
    with open("armazenamento.pkl", "wb") as arquivo:
      pickle.dump(armazenamento, arquivo)

  elif opcao == 2:
    with open("armazenamento.pkl", "rb") as arquivo:
      armazenamento = pickle.load(arquivo)
    print(armazenamento)
  elif opcao == 3:
    print("Saiando!")
    break

  else:
    print("Erro no programa, tente novamente!")



```
